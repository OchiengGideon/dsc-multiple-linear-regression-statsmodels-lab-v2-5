## The Ames Housing Dataset

The [Ames Housing dataset](http://jse.amstat.org/v19n3/decock.pdf) is a newer (2011) replacement for the classic Boston Housing dataset. Each record represents a residential property sale in Ames, Iowa. It contains many different potential predictors and the target variable is `SalePrice`.

````python
import pandas as pd
ames = pd.read_csv("ames.csv", index_col=0)
ames
````

````python
ames.describe()
````

We will focus specifically on a subset of the overall dataset. These features are:

```
LotArea: Lot size in square feet

1stFlrSF: First Floor square feet

GrLivArea: Above grade (ground) living area square feet
```

````python
ames_subset = ames[['LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].copy()
ames_subset
````

## Step 1: Visualize Relationships Between Features and Target

For each feature in the subset, create a scatter plot that shows the feature on the x-axis and `SalePrice` on the y-axis.

````python
# Your code here - import relevant library, create scatter plots
import matplotlib.pyplot as plt
````

````python
# Your written answer here - do these seem like good candidates for linear regression?
fig, ax = plt.subplots(nrows=3, figsize =(10,18))

# first plot: LotArea vs SalePrice
ax[0].scatter(data = ames_subset, x = 'LotArea', y = "SalePrice")
ax[0].set_title('LotArea vs SalePrice')
ax[0].set_xlabel("LotArea")
ax[0].set_ylabel("SalePrice")
# second plot: 1stFlrSF vs SalePrice
ax[1].scatter(data = ames_subset, x = '1stFlrSF', y = "SalePrice")
ax[1].set_title('1stFlrSF vs SalePrice')
ax[1].set_xlabel("1stFlrS")
ax[1].set_ylabel("SalePrice")
# third plot: GrLivArea vs SalePrice
ax[2].scatter(data = ames_subset, x = 'GrLivArea', y = "SalePrice")
ax[2].set_title('GrLivArea vs SalePrice')
ax[2].set_xlabel("GrLivArea")
ax[2].set_ylabel("SalePrice")

plt.tight_layout()
plt.show()
````

## Step 2: Build a Simple Linear Regression Model

Set the dependent variable (`y`) to be the `SalePrice`, then choose one of the features shown in the subset above to be the baseline independent variable (`X`).

Build a linear regression using StatsModels, describe the overall model performance, and interpret its coefficients.

````python
# Your code here - define y and baseline X
y = ames_subset['SalePrice']
baseline_X = ames_subset[['GrLivArea']]
````

````python
# Your code here - import StatsModels, fit baseline model, display results
import statsmodels.api as sm

baseline_model = sm.OLS(y, sm.add_constant(baseline_X))
baseline_model_results = baseline_model.fit()
print(baseline_model_results.summary())
````

````python
# Your written answer here - interpret model results
# R-squared of 0.502 meaning that GrLivArea explains about 50.2% of the variability in SalePrice.
# const of 18,570, means that when GrLivArea is zero, the baseline SalePrice is around 18,570.
# GrLivArea Coefficient of 107.13, means that for each additional square foot in GrLivArea, the SalePrice is predicted to increase by approximately 107.13 while maintainig constants.
````

## Step 3: Build a Multiple Linear Regression Model

For this model, use **all of** the features in `ames_subset`.

````python
# Your code here - define X
x =ames_subset.drop('SalePrice', axis= 1)
x
````

````python
# Your code here - fit model and display results
model=sm.OLS(y, sm.add_constant(x))
model_results = model.fit()
print(model_results.summary())
````

````python
# Your written answer here - interpret model results. Does this model seem better than the previous one?
# R-squared: 0.565, meaning that the model explains about 56.5% of the variance in SalePrice, which is an improvement over using GrLivArea alone
# Intercept: -14,310, suggesting that, with all predictors set to zero, the model would estimate a baseline SalePrice of -14,310, which is unlikely to have practical meaning but indicates the intercept position.
# LotArea: 0.2841, meaning for every additional square foot in LotArea, SalePrice is predicted to increase by about 0.2841 (though its p-value is close to 0.05, so it may have a borderline significance).
# 1stFlrSF: 60.2866, indicating that every additional square foot on the first floor corresponds to an estimated increase of about 60.29 in SalePrice.
# GrLivArea: 80.6061, suggesting a significant positive effect, with each additional square foot increasing SalePrice by about 80.61.
````

## Step 4: Create Partial Regression Plots for Features

Using your model from Step 3, visualize each of the features using partial regression plots.

````python
# Your code here - create partial regression plots for each predictor
fig = plt.figure(figsize=(15,5))
sm.graphics.plot_partregress_grid(model_results,exog_idx=['GrLivArea', '1stFlrSF', 'LotArea'], fig=fig)
plt.tight_layout()
plt.show()
````
